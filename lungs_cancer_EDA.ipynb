```python
#changing gander column  vlaue from object to integer

def change_gender(item):
    if item=='M':
        return 2
    elif item =='F':
        return 1
    
```

```python
df_copy=original_dataframe.copy()
```

```python
df_copy
```

```python
df_copy.GENDER=df_copy.GENDER.apply(change_gender)
```

```python
df_copy
```

```python
#changing lung cancer vlaue from object to integer
def change_lung_cancer(item):
    if item=='YES':
        return 2
    elif item=='NO':
        return 1
```

```python
#changing lung cancer vlaue from object to integer

df_copy.LUNG_CANCER=df_copy.LUNG_CANCER.apply(change_lung_cancer)
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df_copy['AGE'],kde=True,bins=20,color='red')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#lung cancer within the Gender with specific age group
plt.figure(figsize=(15,10))
age_intervals = range(20, 100, 10)

df_copy['AgeInterval'] = pd.cut(df_copy['AGE'], bins=age_intervals, right=False)
lung_cancer_counts = df_copy.groupby(['AgeInterval', 'GENDER'])['LUNG_CANCER'].sum().unstack()
lung_cancer_counts.plot(kind='bar', width=0.35, position=1, color=['blue', 'pink'])
plt.ylabel('Lung Cancer Occurrences')
plt.xlabel('Age Intervals')

plt.legend(title='Gender', labels=['Male', 'Female'])
plt.title('Lung Cancer Occurrences by Age Intervals and Gender')

plt.show()
```

```python

# Histograms and KDE plots
for col in df_copy.columns:
    sns.histplot(df_copy[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
# Box plots
for col in df_copy.columns:
    sns.boxplot(y=df_copy[col])
    plt.title(f'Box plot of {col}')
    plt.show()
```

```python
# Pair plots
sns.pairplot(df_copy)
plt.title('Pair plot')
plt.show()
```

```python
# Scatter plots
for col in df_copy.columns:
    if col != 'LUNG_CANCER':
        sns.scatterplot(x=df_copy[col], y=df_copy['LUNG_CANCER'])
        plt.title(f'Scatter plot of {col} vs LUNG_CANCER')
        plt.show()
```

```python
# Distribution of LUNG_CANCER
sns.histplot(df_copy['LUNG_CANCER'], kde=True)
plt.title('Distribution of LUNG_CANCER')
plt.show()
```

```python
# Compare distributions across target variable
for col in df_copy.columns:
    if col != 'LUNG_CANCER':
        sns.boxplot(x='LUNG_CANCER', y=df_copy[col], data=df_copy)
        plt.title(f'{col} distribution by LUNG_CANCER')
        plt.show()
```
