```python
# Drop columns with only null values
df.drop(columns=['job_type', 'company_type'], inplace=True)

# Check for any other missing values
print(df.isnull().sum())
```

```python
# Summary statistics for numerical columns
print(df[['age', 'weight1']].describe())

# Summary statistics for categorical columns
for col in df.select_dtypes(include='object').columns:
    print(df[col].value_counts())
    print("\n")
```

```python
# Histogram for age
plt.figure(figsize=(10, 5))
sns.histplot(df['age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
df.info()
```

```python
# Bar plot for sex
plt.figure(figsize=(10, 5))
sns.countplot(x='sex', data=df)
plt.title('Distribution of Sex')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

```python
# Bar plot for knows_election_date
plt.figure(figsize=(10, 5))
sns.countplot(x='knows_election_date', data=df)
plt.title('Awareness of Election Date')
plt.xlabel('Knows Election Date')
plt.ylabel('Count')
plt.show()
```

```python
# Bar plot for will_vote
plt.figure(figsize=(10, 5))
sns.countplot(x='will_vote', data=df)
plt.title('Intention to Vote')
plt.xlabel('Will Vote')
plt.ylabel('Count')
plt.show()
```

```python
# Box plot for age vs will_vote
plt.figure(figsize=(10, 5))
sns.boxplot(x='will_vote', y='age', data=df)
plt.title('Age Distribution by Voting Intention')
plt.xlabel('Will Vote')
plt.ylabel('Age')
plt.show()
```

```python
# Pair plot to see relationships between numerical features and will_vote
sns.pairplot(df[['age', 'weight1', 'will_vote']], hue='will_vote')
plt.show()
```
